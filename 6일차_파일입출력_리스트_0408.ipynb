```python
with open('학생 점수.csv', 'r', encoding='utf-8-sig') as f:
    lines = f.readlines()

list_sum = []
list_avg = []
list_b = []

header = lines[0].strip()
list_header = header.split(',')

list_header.append('총합')
list_header.append('평균')


print(list_header)

header = ','.join(map(str, list_header))

for i in range(1, len(lines)):
    b = (lines[i].strip())

    list_b.append(b)

    a = b.split(',')[1:]
    
    a = list(map(int, a))
    
    list_sum.append(sum(a))
    # print(list_sum)


for i in range(0, len(list_sum)):
    list_avg.append(list_sum[i]/len(a))
    # print(list_avg)

print(len(list_sum))
print(len(list_avg))
print(len(lines))

with open('학생점수2.csv', 'w', encoding = 'utf-8-sig') as f:
    f.write(f"{str(header)}\n")
    for i in range(0, len(lines)-1):
        f.write(f"{str(list_b[i])}")
        f.write(f",{str(list_sum[i])}")
        f.write(f",{str(list_avg[i])}\n")

    
    
```

```python
with open('학생 점수.csv', 'r', encoding='utf-8-sig') as f:
  lines = f.readlines()

# 해더
header = lines[0].strip().split(',')
header.extend(['총합', '평균'])
# print(header)

# 해더를 제외한 데이터
data = lines[1:]
new_data = [] # 새로운 데이터를 저장 할 리스트

for line in data: # 요소를 반복
  row = line.strip().split(',')
  scores = row[1:] # 점수만 선택 (학생1, 학생2 등의 내용을 뻄)
  
  total = 0
  for score in scores:
    total += int(score) # int 로 변환하면서 총합 계산
  
  average = total / (len(scores)) # len(scores)는 과목의 개수
  format_average = f'{average:.2f}' # 소숫점 둘째자리까지 반올림
  
  row.extend([str(total), format_average]) # row에 추가(문자로 변경해야함)

  new_data.append(row) # 새로운 리스트에 추가

# print(new_data)

with open('new_학생 점수.csv', 'w', encoding='utf-8-sig') as f:
  f.write(','.join(header) + '\n') # csv는 쉼표로 구문한 데이터, 맨 마지막에 개행문자가 필요함
  for row in new_data:
    f.write(','.join(row) + '\n')
```

```python
def add_num(a, b):
    return a + b, a-b, a*b, a/b
# 함수에 여러개의 변환값을 넣어도 하나로 묶여서 나온다

a = add_num(5, 3)
print(a)

# 함수를 주고 받는 과정에서 튜플 값이 나온다
```

```python
# *를 매개변수로 받을때 관례적으로 args를 같이 써준다
# 이때는 매개 변수가 튜플로 들어가게 된다

def sum_num(*args):
    print(args)
    return sum(args)
# 넣어준 값들이 튜플로 묶여서 들어감
# (10, 20, 30)
# 이후 모든 값들을 더해서 출력

a = sum_num(10, 20, 30)
print(a)
```

```python
t1 = (1, 2, 3, 4, 5)

for i in t1:
    if i % 2 == 0:
        print(i)
        break

def find_first(*args):
    for i in args:
        if i % 2 == 0:
            return i
    return None
    # 가독성과 확실하게 하기 위해 None을 써준다

    
print(find_first(1, 2, 3, 4, 5))
# print(find_first(t1))
# 이 경우에는 t1의 값으로 주다 보니 한개를 반환 한 것으로 판단된다
# 그래서 튜플로 주기 위해 이와 같이 함

# print(find_first(1, 3, 5))
# 만약 2가 없다면? 결과값은 none이다.
```

```python
# 정수로 이루어진 숫자들을 매개변수로 받아 해당 숫자를 한글의 기본 자음으로 반환 하는 함수를 작성하세요(0=ㄱ, 1=ㄴ, 2=ㄷ.....)

list_hangul = ['ㄱ', 'ㄴ', 'ㄷ', 'ㄹ', 'ㅁ', 'ㅂ', 'ㅅ', 'ㅇ', 'ㅈ', 'ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ']

a = int(input("입력하시오"))

def translateHangul(num):
    if num >= len(list_hangul):
        return "다시 확인해주세요"
    elif num < -len(list_hangul):
        return "다시 확인해주세요"
    else:
        return list_hangul[num]

print(translateHangul(a))
```
