```python
merged_data = pd.merge(customer_m, campaign_m, on='campaign_id', how='outer')
merged_data = pd.merge(merged_data, class_m, on='class', how='outer')
merged_data = pd.merge(merged_data, use_l, on='customer_id', how='outer')

# 통합된 데이터프레임 확인
# merged_data.to_csv("통합된데이터.csv")

merged_data = merged_data[['customer_id', 'name', 'class', 'gender', 'start_date', 'end_date', 'usedate', 'log_id', 'campaign_id', 'campaign_name', 'class_name', 'price']]
merged_data
```

```python
def get_merged_customer_id_info(customer_id):
    temp = merged_data[merged_data['customer_id'] == customer_id]

    temp = temp[['customer_id', 'name', 'class', 'usedate', 'price']]
    # print(temp)
    
    customer_id = temp['customer_id'].iloc[0]
    
    # name은 그룹별로 처음 값 선택
    name = temp['name'].iloc[0]

    # 평균 price 계산
    avg_price = temp['price'].mean()

    # 총 price 계산
    total_price = temp['price'].sum()

    # 그룹별로 사용일수(usedate)의 개수(count) 계산
    # class_counts = temp.groupby('customer_id')['usedate'].count()
    class_counts = temp['usedate'].count()
    
    # 출력
    print("ID:", customer_id)
    print("고객명:", name)
    print("평균 이용요금:", round(avg_price))
    print("총 이용요금:", total_price)
    print("사용일수:", class_counts)
    
    # class_counts = temp.groupby('customer_id')['usedate'].count()
    # print(class_counts)

get_merged_customer_id_info('TS999855')
```

```python
def get_merged_day_info(input_day):
    # 각 고객별 사용일수 계산
    customer_usage = merged_data.groupby('customer_id')['usedate'].count()
    
    # 사용일수가 입력받은 값보다 큰 고객 ID 선택
    selected_customers = customer_usage[customer_usage >= input_day].index.tolist()
    
    # 선택된 고객들에 대한 정보 출력
    temp = merged_data[merged_data['customer_id'].isin(selected_customers)]
    temp = temp[['customer_id', 'name', 'class', 'usedate', 'price']]
    print(temp)
    
    
get_merged_day_info(55)
```
