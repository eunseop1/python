```python
T = input("괄호 입력")

stack = []

for i in list(T):
    print(i)
    if i == '(':
        stack.append(i)
        print("if=( ", stack)
    elif i == ')':
        if len(stack) > 0:
            stack.pop()
            print("stack.pop ", stack)
        else:
            print("NO")
            break

if len(stack) == 0:
    print("YES")
    print("yes ",stack)

else:
    print("NO")
```
